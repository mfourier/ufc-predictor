## 3. Load Models

```python
model_list = []

for name in pretty_model_name:
    # Load normal model
    model_normal = UFCModel(model=load_model(name, verbose=True))
    model_list.append(model_normal)

    # Load _no_odds model
    no_odds_name = f"{name}_no_odds"
    try:
        model_no_odds = UFCModel(model=load_model(no_odds_name, verbose=True))
        model_list.append(model_no_odds)
    except FileNotFoundError:
        logger.warning(f"‚ö†Ô∏è No '_no_odds' model found for '{name}', skipping.")
```

## 4. Evaluate Models

```python
for model in model_list:
    model.metrics = evaluate_metrics(model, ufc_data)
    model.cm = evaluate_cm(model, ufc_data)
```
