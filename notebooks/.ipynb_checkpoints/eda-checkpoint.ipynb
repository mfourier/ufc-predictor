## 1. Import Libraries and Setup Environment

```python
# Import necessary libraries
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the project root directory
project_root = '/home/mfourier/Documentos/ufc_predictor'
sys.path.append(os.path.join(project_root, 'data/random'))
import random_data
```

## 2. Load Data

```python
# Load data
df_train, df_test = random_data.random_ufc_data()

# Separate training and testing sets
X_train = df_train.iloc[:, :-1]  # Todas las columnas excepto la Ãºltima
y_train = df_train.iloc[:, -1]   # Solo la Ãºltima columna

X_test = df_test.iloc[:, :-1]
y_test = df_test.iloc[:, -1]
```

## 4. Target Variable Exploration

### Visualizing the Distribution of the Target Variable

```python
# Distribution of the target variable (assuming it's binary)
sns.countplot(x=y_train)
plt.title("Distribution of Target Variable")
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```

## 5. Univariate Analysis of Features (X)

```python
# Histograms of the predictor features
X_train.hist(bins=15, figsize=(15, 10))
plt.suptitle("Distribution of Predictor Features")
plt.show()
```

## 6. Correlation Between Features

```python
# Correlation between features
correlation_matrix = X_train.corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

## 7. Relationships Between Features and Target Variable

```python
# Relationship between features and target variable
for col in X_train.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=y_train, y=X_train[col])
    plt.title(f"Relationship Between {col} and Target Variable")
    plt.xlabel("Class")
    plt.ylabel(col)
    plt.show()
```

## 8. Outliers Analysis

```python
# Boxplots to detect outliers
for col in X_train.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(X_train[col])
    plt.title(f"Outlier Detection for {col}")
    plt.show()
```
